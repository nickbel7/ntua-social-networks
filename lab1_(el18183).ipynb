## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
#n=4x μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
n=43
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG = nx.watts_strogatz_graph(n, d, 0.0)
REG.graph['name'] = 'REG'

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG, node_size=300, node_color="orange", with_labels=True)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
#n=4x 
n=43
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M)
RGER.graph['name'] = 'RGER'

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER, node_size=250, node_color="orange", with_labels=True)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
#n=4x
n=43
R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R, 2)
RGG.graph['name'] = 'RGG'

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
pos = nx.get_node_attributes(RGG, "pos")
dmin = 1
ncenter = 0

for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d
# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

labels={}
for n in RGG.nodes():
  labels[n] = n

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=250,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)
nx.draw_networkx_labels(RGG, pos, labels, font_size=10, font_color="white")

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

## B. Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
def cc(G):
    for n in G :
      G.nodes[n]['cc'] = nx.clustering(G, n)
    #produce histogram
    plt.hist([x[1] for x in G.nodes(data='cc')], label=G.graph['name'], align='left')

# Finds min clustering coefficient in graph G
def minV(G):
    # min_cc = min(list(G.nodes(data='cc')), key=lambda x : x[1])
    min_cc = min(x[1] for x in G.nodes(data='cc'))
    return (min_cc, [x[0] for x in list(G.nodes(data='cc')) if x[1]==min_cc])
  
# Finds max clustering coefficient in graph G
def maxV(G):
    # max_cc = max(list(G.nodes(data='cc')), key=lambda x : x[1])
    max_cc = max(x[1] for x in G.nodes(data='cc'))
    return (max_cc, [x[0] for x in list(G.nodes(data='cc')) if x[1]==max_cc])


#-------------------------------------------------------------------------
Graphs = [REG, RGER, RGG, SF, SW]
# B 1.1
plt.title('B 1.1 | Clusterring coefficient distribution')
for g in Graphs:
  cc(g)

plt.legend(loc='upper left')
plt.show()

#-------------------------------------------------------------------------
print()
# B 1.2
print('B 1.2 | Average Clustering Coefficient')
for g in Graphs :
  print(g.graph['name'] + ' : Mean : ' + str(round(nx.average_clustering(g),4)) + ', Variance : ' + str(round(np.var([x[1] for x in g.nodes(data='cc')]),4)))

#-------------------------------------------------------------------------
print()
# B 1.3
print('B 1.3 | Vertices with MIN & MAX Clustering Coefficient')
for g in Graphs :
  print(g.graph['name'] + ' -> MIN = ' + str(round(minV(g)[0], 2)) + ' : ' + str(minV(g)[1]) + ', ΜΑΧ = ' + str(round(maxV(g)[0], 2)) + ': ' + str(maxV(g)[1]))
```

###Σχολιασμός
- Τοπολογία **REG** : Επειδή όλοι οι κόμβοι της τοπολογίας παρουσιάζουν ακριβώς τα ίδια χαρακτηριστικά τότε όλοι θα πρέπει να έχουν το ίδιο clustering coefficient, το οποίο για d=4 προκύπτει να είναι ίσο με 0.5.
- Τοπολογία **RGER** : Οι συντελεστές ομαδοποίησης είναι σχετικά χαμηλοί και άρα και ο μέσος συντελεστής ομαδοποίησης. Άρα είναι μία τοπολογία χωρίς ιδιαίτερα πολλά clusters.
- Τοπολογία **RGG** : Οι συντελεστές ομαδοποίησης για όλους τους κόμβους είναι πάνω από το 0.5 και έχει και το μεγαλύτερο μέσο συντελεστή. Επομένως, παρατηρούμε ότι είναι η τοπολογία με τα περισσότερα clusters, το οποίο είναι λογικό γιατί βασίζεται τόσο ο συντελεστής ομαδοποίησης, όσο και οι βαθμοί των κόμβων βασίζονται σε τοπολογικά χαρακτηριστικά.
- Τοπολογία **SF** : Έχει σχετικά χαμηλούς συντελεστές για τους περισσότερους κόμβους, αλλά και κάποιες μεγαλύτερες τιμές για κάποιους λιγότερους κόμβους. Αυτό είναι λογικό γιατί παρομοιάζει ένα πραγματικό κοινωνικό δίκτυο οποίο βασίζεται στην αρχή του "ο πλούσιος γίνεται πλουσιότερος".
- Τοπολογία **SW** : Η πλειοψηφία των κόμβων έχουν μικρούς συντελεστές (γιαυτό έχει και μικρό μέσο συντελεστή), ενώ υπάρχουν κάποιοι κόμβοι που έχουν πολύ μεγάλο συντελεστή ομαδοποίησης. Όπως και η Scale-Free, η Small-World είναι πιο κοντινό σαν μοντέλο για πραγματικά κοινωνικά δίκτυα, όπου υπάρχουν λίγοι κόμβοι με μεγάλη επιρροή και περισσότερη με μικρή επιρροή.

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
def compute_path_lengths(G):
    return nx.average_shortest_path_length(G)

# B 2.1
print('B 2.1 | Average Shortest Path')
for g in Graphs :
  if nx.is_connected(g):
    print(g.graph['name'] + ' : ' + str(round(compute_path_lengths(g), 3)))
  else :
    print(g.graph['name'] + ' : Not Connected')

#--------------------------------------------------------------------
print()
# B 2.2
print('B 2.2 | Diameter & Radius')
for g in Graphs :
  if nx.is_connected(g):
    print(g.graph['name'] + ' : ' + 'Diameter = ' + str(nx.diameter(g)) + ', Radius = ' + str(nx.radius(g)))
  else :
    print(g.graph['name'] + ' : Not Connected')
```

###Σχολιασμός
- Τοπολογία **REG** : Παρουσιάζει μεγάλη τοπικότητα στους κόμβους. Κάθε ένας είναι συνδεδεμένος μόνο με κοντινούς του, οπότε συγκριτικά με τις άλλες τοπολογίες περιμέναμε το μεγαλύτερο μέσο όρο μικρότερου μονοπατιού, όπως επίσης και τη μεγαλύτερη διάμετρο, ακτίνα (όντως έχει διάμετρο, ακτίνα 11).
- Τοπολογία **RGER** : Ο γράφος δεν είναι connected, επομένως όλες οι συναρτήσεις επιστρέφουν error, δηλαδή δεν είναι μετρήσιμες αυτές οι μετρικές.
- Τοπολογία **RGG** : Για αυτή τη τοπολογία μπορούμε να ξέρουμε από πριν την ακτίνα και την διάμετρο η οποία θα είναι ίδια με αυτή που ορίσαμε όταν το φτιάξαμε (δηλαδή 3.5->3). Επειδή υπάρχουν πολλές αποστάσεις κοντά στο μήκος της ακτίνας (3), περιμένουμε και το average shortest path να είναι κάπου εκεί κοντά.
- Τοπολογία **SF** : Λόγω των κεντρικών κόμβων, τα scale free έχουνε μικρό average shortest path length, γιατί επιτρέπου μεγάλα άλματα.
- Τοπολογία **SW** : Αντίστοιχα με τα SF, τα small world εχουν και αυτά μικρή σχετικά διάμετρο, ακτίνα αφού μπορούν να εκτελέσουν μεγάλα άλματα από ένα κόμβο σε έναν άλλο.

Συνοπτικά, βλέπουμε ότι το **Average Shortest Path Length** σχετίζεται σε αναλογία με τις ιδιότητες **Diameter** και **Radius** της κάθε τοπολογίας.

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def diam_path_count(paths, d):
    count = 0
    for p in paths :
      if len(paths.get(p)) == d:
        count += 1
    return count

def find_nodes(G, d):
    # print(max(nx.all_pairs_shortest_path(G), key=lambda x : diam_path_count(x[1], d)))
    return max(nx.all_pairs_shortest_path(G), key=lambda x : diam_path_count(x[1], d))

#-------------------------------------------------------------------------------
print('B 2.3 | Node with most diameter shortest paths')
for g in Graphs:
  if nx.is_connected(g):
    print(g.graph['name'] + ' : ')
    d = nx.diameter(g)
    node = find_nodes(g, d)
    print('Node label : ' + str(node[0]))
    print('Nodes : ' + str([x for x in node[1] if len(node[1].get(x)) == d]))
  else :
    print(g.graph['name'] + ' : Not Connected')
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    return nx.periphery(G)
```

```python
def find_center(G):
    return nx.center(G)
```

```python
print('B 2.4 | Periphery & Center')
for g in Graphs :
  if nx.is_connected(g):
    print(g.graph['name'] + ' : ')
    print('Graph Periphery : ' + str(find_periphery(g)))
    print('Graph Center : ' + str(find_center(g)))
  else :
    print(g.graph['name'] + ' : Not Connected')
```

- Τοπολογία RG (ER) : Ο γράφος δεν είναι συνεκτικός, επομένως δεν μπορεί να οριστεί περιφέρεια και κέντρο. (Για οποιεσδήποτε τιμές seed και n έβαλα, ακόμα και seed=1,n=44, ο γράφος δεν γινόταν connected)
- Τοπολογία RGG : Όπως και συμβαίνει με τα αποτελέσματα, περιμένουμε το κέντρο του γράφου να είναι κάποιος από τους κόμβους που βρίσκονται στο κέντρο και η περιφέρεια να είναι οι πιο απομακρυσμένοι κόμβοι, λόγω της ιδιότητας του γράφου η οποία θέτει περιορισμό στο να δημιουργούνται μεγάλα άλματα.
- Τοπολογία SF : Παρατηρούμε ότι το κέντρο του γράφου αποτελείται από κόμβους με μεγάλη επιρροή, ενώ η περιφέρεια από κόμβους με μικρότερο βαθμό.
- Τοπολογία SW : Όπως και στο δίκτυο SF το κέντρο του γράφου είναι ο κόμβος 25, ο οποίος φαίνεται να έχει το μεγαλύτερο βαθμό, ενώ στη περιφέρεια ανήκουν πολλοί κόμβοι που δεν φαίνεται να έχουν τόσο μεγάλο βαθμό και άρα θα χρειάζονται περισσότερα βήματα για να φτάσουν σε όλους τους άλλους κόμβους.

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_degree(G):
    node_centr = nx.degree_centrality(G)
    node_centr_list = [node_centr.get(x) for x in node_centr]
    plt.hist(node_centr_list, label=G.graph['name'], align='left') # Γ 1.1
    print('Mean value : ' + str(round(np.mean(node_centr_list), 3)) + ', Variance : ' + str(round(np.var(node_centr_list), 3))) # Γ 1.2

print('Γ 1.2 | Mean & Variance')
for g in Graphs:
  print(g.graph['name'] + ' : ')
  find_degree(g)

plt.title('Γ 1.1 | Node Degree Centrality')
plt.legend(loc='upper left')
plt.show()
```

###Σχολιασμός
- Τοπολογία REG : Όλοι οι κόμβοι έχουν τον ίδιο βαθμό, άρα και το ίδιο degree centrality
- Τοπολογία RG (ER) : Οι τιμές βαθμών κόμβων είναι σχετικά μικρές, καθώς παρατηρούμε ότι δεν δημιουργούνται κόμβοι με μεγάλους βαθμούς (κόμβοι επιρροής)
- Τοπολογία RGG : Είναι η τοπολογία η οποία παρουσιάζει μεγάλους βαθμούς κόμβου για πολλούς από τους κόμβους, λόγω της σχετικά μεγάλης ακτίνας σαν παράμετρο η οποία επιτρέπει σε κάθε κόμβο να συνδέεται με πολλούς κοντινούς του.
- Τοπολογία SF : Είναι ένας γράφος που περιέχει λίγους κόμβους με μεγάλο βαθμό και πολλούς με μικρότερο. Το ίδιο αποδεικνύει και το ιστόγραμμα (κόκκινο) για τους βαθμούς κόμβων.
- Τοπολογία SW : Ακολουθεί λίγο πολύ την ίδια λογική με τα SF, όμως παρατηρούμε ότι λόγω των μικρών σχετικά παραμέτρων d και p ότι υπάρχει πολύ μικρή διακύμανση από τη μέση τιμή και άρα σχεδόν όλοι οι κόμβοι έχουν μια μικρή κεντρικότητα βαθμού.

### Γ2. Κεντρικότητα εγγύτητας (Closeness centrality)
##### Γ2.1 Να αναπαρασταθεί η κατατανομή της εγγύτητας κόμβων για κάθε τοπολογία.
##### Γ2.2 Να υπολογιστεί η μέση εγγύτητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_closeness(G):
    node_centr = nx.closeness_centrality(G)
    node_centr_list = [node_centr.get(x) for x in node_centr]
    plt.hist(node_centr_list, label=G.graph['name'], align='left') # Γ 2.1
    print('Mean value : ' + str(round(np.mean(node_centr_list), 3)) + ', Variance : ' + str(round(np.var(node_centr_list), 3))) # Γ 2.2

print('Γ 2.2 | Mean & Variance')
for g in Graphs:
  print(g.graph['name'] + ' : ')
  find_closeness(g)

plt.title('Γ 2.1 | Node Closeness Centrality')
plt.legend(loc='upper left')
plt.show()
```

###Σχολιασμός
- Τοπολογία REG : Είναι ένας γράφος με μεγάλη μέση απόσταση από ένα κόμβο προς όλους τους άλλους και γιαυτό όλοι οι κόμβοι έχουν μικρή κεντρικότητα εγγύτητας.
- Τοπολογία RG (ER) : Η τυχαιότητα με την οποία φτιάγνεται αυτός ο γράφος του επιτρέπει να υπάρχουν μικρά σχετικά μονοπάτια από κάθε κόμβο σε κάθε άλλο, γιαυτό και βλέπουμε να υπάρχουν πολλοί κόμβοι με μεγάλη τιμή κεντρικότητας εγγύτητας.
- Τοπολογία RGG : Σε αυτή τη τοπολογία υπάρχουν κάποιοι κόμβοι που βρίσκονται σε κεντρικό σημείο του γράφου και αυτοί θα έχουν και τους μεγαλύτερους συντελεστές. Ακόμα όμως, υπάρχουν και πολλοί οι οποίοι βρίσκονται στη περιφέρεια και θα έχουν σχετικά μικρότερους συντελεστές. Καθώς υπάρχει ποικιλία παρατηρείται και ανίστοιχα μεγάλη διακύμανση
- Τοπολογία SF : Εδώ περιμένουμε ότι λόγω της ύπαρξης των κόμβων επιρροής (που έχουν μεγάλο βαθμό), θα ελλατωθούν όλες οι υπόλοιπες ελάχιστες αποστάσεις. Γιαυτό και βλέπουμε στο ιστόγραμμα ότι η μέση τιμή, αλλά και όλοι οι συντελεστές είναι αρκετά μεγάλοι. Φαίνεται να είναι η τοπολογία με τις μεγαλύτερες κεντρικότητες εγγύτητας.
- Τοπολογία SW : Μιάς και δεν υπάρχουν τόσοι πολλοί κόμβοι με μεγάλο βαθμό (λόγω των παραμέτρων d,p) περιμένουμε οι συντελεστές να είναι λίγο μικρότεροι από αυτούς του SF, καθώς δεν θα υπάρχουν τόσα πολλά μεγάλα άλματα που να μειώνουν τη μέση ελάχιστη απόσταση.

### Γ3. Ενδιαμεσική κεντρικότητα (Betweenness centrality)
##### Γ3.1 Να αναπαρασταθεί η κατατανομή της ενδιαμεσικής κεντρικότητας κόμβων για κάθε τοπολογία.
##### Γ3.2 Να υπολογιστεί η μέση ενδιαμεσική κεντρικότητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_betweenness(G):
    node_centr = nx.betweenness_centrality(G)
    node_centr_list = [node_centr.get(x) for x in node_centr]
    plt.hist(node_centr_list, label=G.graph['name'], histtype='step') # Γ 3.1
    print('Mean value : ' + str(round(np.mean(node_centr_list), 3)) + ', Variance : ' + str(round(np.var(node_centr_list), 3))) # Γ 3.2

print('Γ 3.2 | Mean & Variance')
for g in Graphs:
  print(g.graph['name'] + ' : ')
  find_betweenness(g)

plt.title('Γ 3.1 | Node Betweenness Centrality')
plt.legend(loc='upper right')
plt.show()
```

###Σχολιασμός
- Τοπολογία REG : Όλοι οι κόμβοι περιμένουμε να έχουν την ίδια τιμή και αυτή να είναι σχετικά μεγάλη, γιατί ο γράφος είναι σαν αλυσίδα και για να πας από έναν κόμβο σε έναν άλλο το ελάχιστο μονοπάτι είναι περίπου η περιμετρική απόσταση (άρα κάθε φορά περνάμε από πολλούς κόμβους).
- Τοπολογία RG (ER) : Λόγω της τυχαιότητας δεν περιμένουμε να υπάρχουν πολύ μεγάλα μονοπάτια ούτε κόμβοι επιρροής και άρα οι περισσότεροι κόμβοι θα πρέπει να έχουν μικρό betweenness centrality (όπως και συμβαίνει στη κατανομή).
- Τοπολογία RGG : Ο γεωμετρικός γράφος έχει πολλούς κόμβους που βρίσκονται στη περιφέρεια, άρα δεν θα περνάνε πολλά μονοπάτια από αυτούς, αλλά και κάποιους λίγους, πιό κεντρικούς κόμβους, οι οποίοι περιμένουμε να έχουν μεγάλους συντελεστές γιατί θα συνδέουν πολλά clusters μεταξύ τους.
- Τοπολογία SF : Σε ακόμα μεγαλύτερο βαθμό από τον RGG γράφο, ο SF δημιουργεί λίγους κόμβους με μεγάλους βαθμούς, οι οποίοι αυξάνουν έτσι τις πιθανότητες να γίνονται μέρη ελαχίστων μονοπατιών. Άρα και πάλι έχουμε πολλούς κόμβους με μικρούς συντελεστές και λίγους με μεγαλύτερους συντελεστές ενδιαμεσικής κεντρικότητας.
- Τοπολογία SW : Χωρίς να υπάρχουν κόμβοι με τόσο μεγάλους βαθμούς, παρατηρούμε μια μικρότερη αυστηρότητα ώς προς το ποιοι κόμβοι τείνουν να ανήκουν πιο συχνά σε ελάχιστα μονοπάτια. Γιαυτό και η κατανομή είναι μετατοπισμένη δεξιότερα του SF.

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G):
    l_max = np.real(max(nx.adjacency_spectrum(G))) # imaginary part always 0
    alpha_value = 1 / l_max - 0.01
    node_centr = nx.katz_centrality(G, alpha=alpha_value)
    node_centr_list = [node_centr.get(x) for x in node_centr]
    plt.hist(node_centr_list, label=G.graph['name'], align='right') # Γ 4.1
    print('Mean value : ' + str(round(np.mean(node_centr_list), 3)) + ', Variance : ' + str(round(np.var(node_centr_list), 3))) # Γ 4.2

print('Γ 4.2 | Mean & Variance')
for g in Graphs:
  print(g.graph['name'] + ' : ')
  find_katz(g)

plt.title('Γ 4.1 | Node Katz Centrality')
plt.legend(loc='upper left')
plt.show()
```

###Σχολιασμός
- Τοπολογία REG : O κάθε κόμβος έχει ακριβώς τον ίδιο αριθμό γειτόνων όπου ο καθένας έχει την ίδια ακριβώς "επιρροή". Άρα όλοι θα έχουν τον ίδιο συντελεστή ο οποίος προκύπτει να μην είναι ούτε πολύ μικρός, ούτε πολύ μεγάλος.
- Τοπολογία RG (ER) : Ο κάθε κόμβος λόγω της τυχαιότητας του γράφου περιμένουμε λίγο πολύ να έχει τα ίδια χαρακτηριστικά με τους γείτονές του. Άρα και πάλι οι συντελεστές δεν έχουν μεγάλη διακύμανση.
- Τοπολογία RGG : Υπάρχουν κόμβοι οι οποίοι ίσως είναι πιο απομονωμένοι και θα παρουσιάζουν μικρότερη κεντρικότητα katz (βλέπουμε ότι το πράσινο ξεφεύγει προς τα αριστερά). Υπάρχουν όμως και κόμβοι που βρίσκονται σε πυκνά clusters και τότε περιμένουμε να έχουν μεγαλύτερους συντελεστές.
- Τοπολογία SF : Εδώ οι κόμβοι με τη μεγαλύτερη επιρροή και βαθμό έχουν πολλούς γείτονες οι οποίοι αθροιστικά τους δίνουν μεγάλη αξία, άρα και μεγάλο συντελεστή κεντρικότητας. Οι υπόλοιποι κόμβοι που δεν έχουν τοσο μεγάλους βαθμούς έχουν και μικρότερους συντελεστές.
- Τοπολογία SW : Ένα distribution όπως και το SF όμως με λιγότερες ακραίες τιμές και πάλι, περισσότερους κόμβους με μικρούς συντελεστές και λιγότερους με μεγαλύτερες κεντρικότητες.

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G):
    # DEGREE CENTRALITY
    node_centr = nx.degree_centrality(G) # dictionary
    max_n = node_centr.get(max(node_centr, key=lambda x: node_centr.get(x)))
    node_max_list = [x for x in node_centr if node_centr.get(x) == max_n]
    print('Degree Centrality : ' + str(node_max_list))

    # CLOSENESS CENTRALITY
    node_centr = nx.closeness_centrality(G) # dictionary
    max_n = node_centr.get(max(node_centr, key=lambda x: node_centr.get(x)))
    node_max_list = [x for x in node_centr if node_centr.get(x) == max_n]
    print('Closeness Centrality : ' + str(node_max_list))
    
    # BETWEENESS CENTRALITY
    node_centr = nx.betweenness_centrality(G) # dictionary
    max_n = node_centr.get(max(node_centr, key=lambda x: node_centr.get(x)))
    node_max_list = [x for x in node_centr if node_centr.get(x) == max_n]
    print('Betweeness Centrality : ' + str(node_max_list))

    # KATZ CENTRALITY
    l_max = np.real(max(nx.adjacency_spectrum(G))) # imaginary part always 0
    alpha_value = 1 / l_max - 0.01
    node_centr = nx.katz_centrality(G, alpha=alpha_value) # dictionary
    max_n = node_centr.get(max(node_centr, key=lambda x: node_centr.get(x)))
    node_max_list = [x for x in node_centr if node_centr.get(x) == max_n]
    print('Katz Centrality : ' + str(node_max_list))

for g in Graphs : 
  print(g.graph['name'] + ' : ')
  find_central(g)
```

###Σχολιασμός
- Τοπολογία REG : Εδώ προφανώς, όλοι οι κόμβοι θεωρούνται κεντρικοί για όλα τα κριτήρια, αφού έχουν ακριβώς τα ίδια χαρακτηριστικά.
- Τοπολογία RG (ER) : Εδώ έτυχε ο 27 να είναι ένας κόμβος με λίγο μεγαλύρο βαθμό κόμβου, γιαυτό και έχει μία μεγαλύτερη κεντρικότητα.
- Τοπολογία RGG : Εδώ ο 15 έχει και μεγάλο βαθμό κόμβου (μαζί με τον 40), αλλά βρίσκεται και σε κεντρικό σημείο, το οποίο σημαίνει ότι λειτουργεί σαν γέφυρα για πολλά clusters του γράφου (άρα ανήκει και σε πολλά ελάχιστα μονοπάτια κλπ).
- Τοπολογία SF : Από αυτά που έχουμε πει και προηγουμένως για τον SF μεγάλη κεντρικότητα έχουν οι κόμβοι με μεγάλο βαθμό κόμβου. Εδώ ένας τέτοιος είναι και ο κόμβος 0. 
- Τοπολογία SW : Εδώ αν και ισχύει το ίδιο κριτήριο με τον SF, δεν είναι τόσο διακριτή η διαφορά μεταξύ κόμβων με μικρό και μεγάλο βαθμό κόμβου. Γιαυτό και δεν βλέπουμε ένα κόμβο να υπερισχύει στις κεντρικότες, αλλά διαφορετικό για κάθε τύπο κεντρικότητας.

```python
#PAGERANK
#Calculation
ranks = nx.pagerank(G40)
print(ranks)

#Visualization
plt.title("Page Rank web-Stanford Histogram")
plt.hist([ranks.get(x) for x in ranks])
plt.plot

pos = nx.spring_layout(G40, k=1.0) # the bigger the k the furthest the nodes are from each other
# pos = nx.circular_layout(G40)
# pos = nx.shell_layout(G40)

# Graph visualization
node_sizes = []
for n in G40.degree():
  node_sizes.append(10 * n[1]**1.3)

plt.figure(figsize=(10, 10))
nx.draw_networkx_edges(G40, pos, alpha=0.4)
nx.draw_networkx_nodes(
    G40,
    pos,
    nodelist=list(ranks.keys()),
    node_size=node_sizes,
    node_color=list(ranks.values()),
    cmap=plt.cm.Reds,
)

plt.title("Page Rank web-Stanford")
plt.axis("off")
plt.show()
```

### E) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

##### Μεταβάλλοντας την παράμετρο $g_p$ από $0$ έως $1$ με βήμα $0.1$της συνάρτησης που κατασκευάζει τοπολογίες SW μεγέθους $n=30x$ ( αντικαταστήστε το x με το τελευταίο ψηφίο του ΑΜ) με βαθμό αρχικού πλέγματος $d=4$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
random.seed(seedno)
#n=30x
n=303
d=4

median_paths = []
clustering_coeff = []

for p in np.arange(0, 1, 0.1) :
  SW_instance = nx.watts_strogatz_graph(n, d, p) 
  median_paths.append(nx.average_shortest_path_length(SW_instance))
  clustering_coeff.append(nx.average_clustering(SW_instance))

plt.title('Average Path Length')
plt.plot(median_paths)
plt.show()

plt.title('Average Clustering Coefficient')
plt.plot(clustering_coeff, color='orange')
plt.show()
```

###Σχολιασμός
Για τιμή $g_p=0$ το δίκτυο συμπεριφέρεται σαν ένα Regular Graph, ενώ για $g_p=1$ ώς ένα Random Graph. Για αυτό βλέπουμε ότι όσο ο γράφος γίνεται πιο Random, τόσο μικραίνει και το μέσο μήκος μονοπατιού γιατί δημιουργούνται περισσότερα μεγάλα άλματα στο δίκτυο. Για ενδοιάμεσες τιμές (0.1-0.5), οι οποίες τείνουν πιο πολύ προς το 0 παρά προς το 1, το δίκτυο συμπεριφέρεται σαν ένα Small-world δίκτυο ,το οποίο προσεγγίζει και ένα πραγματικό κοινωνικό δίκτυο.
